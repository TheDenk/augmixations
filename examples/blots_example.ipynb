```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from augmixations import HandWrittenBlot
```

##Help functions

```
def show_img(image, figsize=(4, 6), title=None):
    img = image.copy()
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=figsize)
    plt.title(title)
    plt.imshow(img)
    plt.axis(False)
    plt.show()
```

##Create images

```
img_path = './augmixations/images/handwritten_example.png'
image = cv2.imread(img_path)
show_img(image, figsize=(12, 4))
```

##Simple usage

```
blots = HandWrittenBlot(None, None)
new_img = blots(image)

show_img(new_img, figsize=(12, 4))
```

##Advanced usage (with configs)

```
rect_config = {
    'x': (None, None),
    'y': (None, None),
    'h': (None, None),
    'w': (None, None),
}

blot_params = {
    'incline': (-10, 10),
    'intensivity': (0.5, 0.9),
    'transparency': (0.05, 0.4),
    'count': 5,
}
```

```
blots = HandWrittenBlot(rect_config, blot_params)
new_img = blots(image)

show_img(new_img, figsize=(12, 4))
```
